```python
MEDIA_ALIGNMENTS = {"MSNBC" : -0.8, "Late Night" : -0.5, "New York Times" : -0.3, "Fox News" : 0.5, "OAN" : 0.8}
```

```python
@jit((float64[:], int64), nopython=True, nogil=True)
def ewma(arr_in, window):
    n = arr_in.shape[0]
    ewma = np.empty(n, dtype=float64)
    alpha = 2 / float(window + 1)
    w = 1
    ewma_old = arr_in[0]
    ewma[0] = ewma_old
    for i in range(1, n):
        w += (1-alpha)**i
        ewma_old = ewma_old*(1-alpha) + arr_in[i]
        ewma[i] = ewma_old / w
    return ewma
```

```python
class Agent:
    def __init__(self, polling_opinion, internal_opinion):
        #scale -1 to 1
        self.polling_opinion = polling_opinion if abs(polling_opinion) < 1 else np.sign(polling_opinion)
        
        #scale 0 to 1 dependant on pol_alignment
        self.persuadability = 1 - abs(self.polling_opinion)
        
        #opinion of trump, pre-dist -1 to 1
        self.internal_opinion = internal_opinion
        
        #probabilities to consume each media outlet each timestep
            #dictionary 
        self.media_consumption_percentages = MEDIA_ALIGNMENTS.copy()
        self.generate_media_percentages()
        
    
    def print_stats(self):
        print(f'Political Alignment : {self.polling_opinion}')
        print(f'Persuadability : {self.persuadability}')
        print(f'Opinion : {self.internal_opinion}')
        print(f'Media Consumption Percentages : {self.media_consumption_percentages}', flush=True)
        
    
    def generate_media_percentages(self):
        
        for media in MEDIA_ALIGNMENTS.keys():
            #this value is from 0 to 2
            dist = np.log((-1+np.e**2)*0.5*abs(self.polling_opinion - MEDIA_ALIGNMENTS[media])+1)
            self.media_consumption_percentages[media] = (100 - dist*50)/100
       
            #add noise
            #self.media_consumption_percentages[media] += np.random.rand()*20 - 10
            
            #make percentage
            #self.media_consumption_percentages[media] /= 100
#             if self.media_consumption_percentages[media] < 0: self.media_consumption_percentages[media] = 0
#             if self.media_consumption_percentages[media] > 1: self.media_consumption_percentages[media] = 1
```

```python
class ABM: # {-1:0.25, -0.33:0.25, 0.33:0.25, 1:0.25}
    def __init__(self, number_of_agents, agent_political_distribution={-1:0.05, -0.5:0.25, 0:0.4, 0.5:0.25, 1:0.05}, partitions=0):
        if partitions>0: assert(len(agent_political_distribution)==partitions)
        
        print("Initializing agents...")
        self.political_distribution = agent_political_distribution
        
        self.make_agents(number_of_agents, agent_political_distribution, partitions)
        
        self.chosen_index = -1
        self.chosen_history = [0, {x : 0 for x in MEDIA_ALIGNMENTS.keys()}, 0]
        
        self.starting_stats = [0, 0]
        print("Done")
        
    #polling is a dict with {political alignment : probability}    
    def make_agents(self, number_of_agents, agent_political_distribution, partitions):
        self.agents = []
        pool = number_of_agents
        
        alignments = list(agent_political_distribution.keys())
        if partitions: parts = {np.round(x, 2) : np.round(x+(1/partitions), 2) for x in np.arange(-1, 1, 1/partitions)}
            
        for i, (political_value, weight) in enumerate(agent_political_distribution.items()):
            num = round(number_of_agents * weight)
            if num > pool: num = pool
            pool -= num
            
            

            try:
                next_val = alignments[i+1]
            except:
                next_val = alignments[i-1]
            
            for _ in range(num):
                #do we want to add noise here?
                    #add some noise to the opinion for now based on pol-dist
                if partitions:
                    a = list(parts.keys())[i]
                    
                    alignment = np.random.normal(a + 1/(partitions*2), 1/(partitions*6)) 
                else:
                    alignment = np.random.normal(political_value, abs(political_value - next_val)/3) 
                    
                    
                if abs(alignment) > 1:
                    alignment = political_value - np.sign(political_value)*abs(np.random.normal(0, 0.2/3))
                    
                opinion = np.random.normal(alignment, (1/len(agent_political_distribution))/3) if abs(political_value) != 1 else alignment - np.sign(alignment)*abs(np.random.normal(0, 0.2/3))

                self.agents.append(Agent(alignment, opinion))

                
                
    
    
    #add decay each timestep to momentary (towards original), polling is just ewma (or maybe some other one)
    
    def step(self, factor, current_step): 
        decay = factor
        for i, agent in enumerate(self.agents):
            
            agent.internal_opinion += (self.history[0, i, 0] - agent.internal_opinion)*decay
            for j, (media_name, percent) in enumerate(agent.media_consumption_percentages.items()):
                
                if np.random.choice([True, False], p=[percent, 1-percent]):
                    #did consume this media
                    if i == self.chosen_index: self.chosen_history[1][media_name] += 1
                        
                    self.media_consumed[current_step, i, j] += 1
#                     if np.sign(MEDIA_ALIGNMENTS[media_name]*agent.internal_opinion) == 1 and abs(agent.internal_opinion) > abs(MEDIA_ALIGNMENTS[media_name]):
#                         continue
                    agent.internal_opinion += agent.persuadability * (MEDIA_ALIGNMENTS[media_name]-agent.internal_opinion) * factor
        
                    
                    
                    if abs(agent.internal_opinion) > 1: agent.internal_opinion = np.sign(agent.internal_opinion)
            
            self.history[current_step, i, 0] = agent.internal_opinion
            
            e = ewma(self.history[:current_step+1, i, 0], 5)
            agent.polling_opinion = e[-1]
            agent.persuadability = 1 - abs(agent.polling_opinion)
            
            if abs(agent.polling_opinion) > 1: agent.internal_opinion = np.sign(agent.internal_opinion)
    
    
    
    def simulate(self, timesteps, draw=True, factor=.01, verbose=True):
        num_agents = len(self.agents)
        self.history = np.zeros((timesteps, num_agents, 1), dtype=float)
        self.media_consumed = np.zeros((timesteps, num_agents, len(MEDIA_ALIGNMENTS)))
        self.chosen_index = np.random.randint(num_agents)
        self.chosen_history[0] = self.agents[self.chosen_index].internal_opinion
        
        self.starting_stats = [100*len([self.agents[i].internal_opinion for i in range(len(self.agents)) if self.agents[i].internal_opinion  > 0])/len(self.agents),
                               100*len([self.agents[i].internal_opinion for i in range(len(self.agents)) if self.agents[i].internal_opinion  < 0])/len(self.agents)]
        
        print("Simulating...")
        
        if draw:
            fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
#             plt.xlim([-1, 1])
#             plt.ylim([-1, 1])
#             plt.xlabel("polling_opinion")
#             plt.ylabel("internal_opinion")
            empty, = plt.plot([])
            ax1.set_xticklabels(MEDIA_ALIGNMENTS.keys(), rotation='vertical')
            bars = ax1.bar(range(len(MEDIA_ALIGNMENTS)), np.sum(self.media_consumed, axis=(1, 0)))
            
            #ln, = plt.plot([self.agents[i].polling_opinion for i in range(len(self.agents))], [self.agents[i].internal_opinion for i in range(len(self.agents))], 'o')
            #chosen_pt, = plt.plot(self.agents[self.chosen_index].polling_opinion, self.agents[self.chosen_index].internal_opinion, 'o')
            
            
            
#             colors = [np.round([np.ceil(agent.polling_opinion)*agent.polling_opinion, .1, np.floor(agent.polling_opinion)*agent.polling_opinion], 3) for agent in self.agents]
            
#             rand = 2*np.random.rand(len(self.agents), 2)-1
#             points = plt.scatter(rand[:, 0], rand[:, 1], c=colors, s=500)

            pos = [np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) for agent in self.agents if np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) > 0]
            neg = [np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) for agent in self.agents if np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) < 0]

            pie = ax2.pie([len(pos), len(neg)], labels=["Approve", "Disapprove"], autopct='%1.1f%%')
            ax2.axis('equal')
            fig.set_facecolor('white')
            
            
            '''Also try to do a subplot with hist of media consumed, maybe all agents on 1D line with horizontals to mark their opinion'''
            
            print('\r[' + '>' + '-'*(19) + ']', end=' ', flush=True) 
            
            
            def update(frame, self):
                self.step(factor, frame)
                #self.history[frame] = np.array([self.agents[j].internal_opinion for j in range(len(self.agents))]).reshape(-1, 1)
                temp = int(frame // (timesteps/20))
                if temp == 0: temp = 1
                print('\r[' + '='*(temp-1) + '>' + '-'*(20 - temp) + ']', end=' ', flush=True) 
                
                '''2D Plot'''
                #ln.set_data([self.agents[i].polling_opinion for i in range(len(self.agents))], [self.agents[i].internal_opinion for i in range(len(self.agents))])
                #chosen_pt.set_data(self.agents[self.chosen_index].polling_opinion, self.agents[self.chosen_index].internal_opinion)
                
                '''Histogram'''
                #ax1.clear()
                totals = np.sum(self.media_consumed, axis=(1, 0))
                for i, bar in enumerate(bars):
                    bar.set_height(totals[i])
                
                ax1.relim()
                ax1.autoscale_view()
                '''Colors'''
                #colors = [np.round([np.ceil(agent.polling_opinion)*agent.polling_opinion, .1, np.floor(agent.polling_opinion)*agent.polling_opinion], 3) for agent in self.agents]
                #points.set_color(colors)
               
            
                pos = [np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) for agent in self.agents if np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) > 0]
                neg = [np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) for agent in self.agents if np.round((2/np.pi)*np.arctan(1000*agent.polling_opinion)) < 0]

                ax2.clear()
                ax2.axis('equal')
                ax2.pie([len(pos), len(neg)], labels=["Approve", "Disapprove"], autopct='%1.1f%%')

                if frame == timesteps-1:
                    print('\r[' + '='*(19) + '>' + '-'*(0) + ']', end='', flush=True) 
                    print('\nDone!')
                
                return (*(bars)), empty
            
            anim = animation.FuncAnimation(fig, update, frames=range(timesteps), fargs=(self, ), interval=200, blit=True)
            display(HTML(anim.to_html5_video()))
     
        else:
            print('\r[' + '>' + '-'*(19) + ']', end=' ', flush=True) 
            cur = 0
            for i in range(timesteps):
                #self.history[i] = np.array([self.agents[j].internal_opinion for j in range(num_agents)]).reshape(-1, 1)

                if int(i // (timesteps/20)) > cur:
                    cur = int(i // (timesteps/20))
                    print('\r[' + '='*(cur+1) + '>' + '-'*(19 - cur) + ']', end=' ', flush=True) 
                self.step(factor, i)
                print('\r[' + '='*(19) + '>' + '-'*(0) + ']', end='', flush=True) 
                print('\nDone!')
            

        
        self.chosen_history[-1] = self.agents[self.chosen_index].internal_opinion
        if verbose:
            print(f'\nAgent {self.chosen_index}:\n\tOpinion Change: {self.chosen_history[0]} -> {self.chosen_history[-1]}\n\tMedia Percentages: {[name + ": {}%".format(int(1000*percent*100)//1000) for name, percent in self.agents[self.chosen_index].media_consumption_percentages.items()]}\n\tMedia Consumed: {self.chosen_history[1]}')
            self.stats()
        
        if draw:
            print("\nDrawing...")
            #river plot
            self.river_plot(num_agents, timesteps)
    
    
    
    
    
    def river_plot(self, num_agents, timesteps):
        plt.figure(figsize=(20, 10))
        for agent in range(num_agents):
            plt.plot(np.arange(timesteps), self.history[:, agent, 0], linewidth=100/num_agents)
        plt.xlabel('Timesteps')
        plt.ylabel('Opinion')
        plt.show()
        
    def stats(self):
        print(f'\nOverall Approval: {self.starting_stats[0]}% -> {100*len([abm.agents[i].internal_opinion for i in range(len(abm.agents)) if abm.agents[i].internal_opinion  > 0])/len(abm.agents)}%')
        print(f'Overall Disapproval: {self.starting_stats[1]}% -> {100*len([abm.agents[i].internal_opinion for i in range(len(abm.agents)) if abm.agents[i].internal_opinion  < 0])/len(abm.agents)}%')

    
    def draw(self, save=False):
        opinions = [self.agents[i].internal_opinion for i in range(len(self.agents))]
        political_align = [self.agents[i].polling_opinion for i in range(len(self.agents))]
        plt.scatter(political_align, opinions)
        plt.title("Starting Distribution")
        plt.xlabel("Politcal Alignment")
        plt.ylabel("Opinion of Trump")
        
        if save: plt.savefig("Starting_Distribution")
        plt.show()    
    
    
```

```python
import numpy as np
from numba import jit
from numba import float64
from numba import int64


@jit((float64[:], int64), nopython=True, nogil=True)
def _ewma(arr_in, window):
    r"""Exponentialy weighted moving average specified by a decay ``window``
    to provide better adjustments for small windows via:

        y[t] = (x[t] + (1-a)*x[t-1] + (1-a)^2*x[t-2] + ... + (1-a)^n*x[t-n]) /
               (1 + (1-a) + (1-a)^2 + ... + (1-a)^n).

    Parameters
    ----------
    arr_in : np.ndarray, float64
        A single dimenisional numpy array
    window : int64
        The decay window, or 'span'

    Returns
    -------
    np.ndarray
        The EWMA vector, same length / shape as ``arr_in``

    Examples
    --------
    >>> import pandas as pd
    >>> a = np.arange(5, dtype=float)
    >>> exp = pd.DataFrame(a).ewm(span=10, adjust=True).mean()
    >>> np.array_equal(_ewma_infinite_hist(a, 10), exp.values.ravel())
    True
    """
    n = arr_in.shape[0]
    ewma = np.empty(n, dtype=float64)
    alpha = 2 / float(window + 1)
    w = 1
    ewma_old = arr_in[0]
    ewma[0] = ewma_old
    for i in range(1, n):
        w += (1-alpha)**i
        ewma_old = ewma_old*(1-alpha) + arr_in[i]
        ewma[i] = ewma_old / w
    return ewma


@jit((float64[:], int64), nopython=True, nogil=True)
def _ewma_infinite_hist(arr_in, window):
    r"""Exponentialy weighted moving average specified by a decay ``window``
    assuming infinite history via the recursive form:

        (2) (i)  y[0] = x[0]; and
            (ii) y[t] = a*x[t] + (1-a)*y[t-1] for t>0.

    This method is less accurate that ``_ewma`` but
    much faster:

        In [1]: import numpy as np, bars
           ...: arr = np.random.random(100000)
           ...: %timeit bars._ewma(arr, 10)
           ...: %timeit bars._ewma_infinite_hist(arr, 10)
        3.74 ms ± 60.2 µs per loop (mean ± std. dev. of 7 runs, 100 loops each)
        262 µs ± 1.54 µs per loop (mean ± std. dev. of 7 runs, 1000 loops each)

    Parameters
    ----------
    arr_in : np.ndarray, float64
        A single dimenisional numpy array
    window : int64
        The decay window, or 'span'

    Returns
    -------
    np.ndarray
        The EWMA vector, same length / shape as ``arr_in``

    Examples
    --------
    >>> import pandas as pd
    >>> a = np.arange(5, dtype=float)
    >>> exp = pd.DataFrame(a).ewm(span=10, adjust=False).mean()
    >>> np.array_equal(_ewma_infinite_hist(a, 10), exp.values.ravel())
    True
    """
    n = arr_in.shape[0]
    ewma = np.empty(n, dtype=float64)
    alpha = 2 / float(window + 1)
    ewma[0] = arr_in[0]
    for i in range(1, n):
        ewma[i] = arr_in[i] * alpha + ewma[i-1] * (1 - alpha)
    return ewma
```
